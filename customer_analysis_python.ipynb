```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace("purchase_amount_(usd)","purchase_amount")
df.columns
```

```python
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
freequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_freequency_days'] = df['frequency_of_purchases'].map(freequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_freequency_days']].head(20)
```

```python
df = df.drop('promo_code_used', axis =1)
```
